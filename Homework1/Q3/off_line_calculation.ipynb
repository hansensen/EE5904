```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([-1,1,3])
y = x * -1 + 1.5
f, ax = plt.subplots(1)
plt.plot(x, y)
plt.scatter(1,1, c='b', marker='^', s=100)
plt.scatter(0,0, c='r', s=100)
plt.scatter(0,1, c='r', s=100)
plt.scatter(1,0, c='r', s=100)
plt.ylabel('x')
plt.xlabel('y')
ax.set_ylim(ymin=-0.5, ymax=2.5)
ax.set_xlim(xmin=-0.5, xmax=2.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.savefig('q3and.png')
plt.show()
```
